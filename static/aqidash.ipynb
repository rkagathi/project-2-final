```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
bdate = "20180601"
edate = "20180630"

pmurl = f"https://aqs.epa.gov/api/rawData?user=thomas.e.abraham@gmail.com&pw=khakiosprey52&format=AQCSV&param=81102&bdate={bdate}&edate={edate}&state=17&county=031"
ozurl = f"https://aqs.epa.gov/api/rawData?user=thomas.e.abraham@gmail.com&pw=khakiosprey52&format=AQCSV&param=44201&bdate={bdate}&edate={edate}&state=17&county=031"
courl = f"https://aqs.epa.gov/api/rawData?user=thomas.e.abraham@gmail.com&pw=khakiosprey52&format=AQCSV&param=42101&bdate={bdate}&edate={edate}&state=17&county=031"
sourl = f"https://aqs.epa.gov/api/rawData?user=thomas.e.abraham@gmail.com&pw=khakiosprey52&format=AQCSV&param=42401&bdate={bdate}&edate={edate}&state=17&county=031"
nourl = f"https://aqs.epa.gov/api/rawData?user=thomas.e.abraham@gmail.com&pw=khakiosprey52&format=AQCSV&param=42602&bdate={bdate}&edate={edate}&state=17&county=031"

pmdata = pd.read_csv(pmurl)
ozdata = pd.read_csv(ozurl)
codata = pd.read_csv(courl)
sodata = pd.read_csv(sourl)
nodata = pd.read_csv(nourl)
```

```python
new = pd.concat([pmdata,ozdata], ignore_index=True)
newer = pd.concat([new,codata], ignore_index=True)
newest = pd.concat([newer,sodata], ignore_index=True)
june = pd.concat([newest,nodata], ignore_index=True)
```

```python
june.site.unique()
june.lat.unique()
june.lon.unique()

location_coords = pd.DataFrame(data={'id':june.site.unique(), 'x':june.lat.unique(), 'y':june.lon.unique()})
location_coords.set_index('id')
```

```python
location_coords=location_coords.drop(location_coords.index[3])

location_coords=location_coords.set_index('id')

location_coords.to_csv('data/location_coords.csv')
```

```python
oz = ozdata[['site','datetime','value']]
oz = oz.rename(index=str, columns={"value":"o3"})
co = codata[['site','datetime','value']]
co = co.rename(index=str, columns={"value":"co"})
so = sodata[['site','datetime','value']]
so = so.rename(index=str, columns={"value":"so2"})
no = nodata[['site','datetime','value']]
no = no.rename(index=str, columns={"value":"no2"})
pm = pmdata[['site','datetime','value']]
pm = pm.rename(index=str, columns={"value":"pm10"})
```

```python
values = oz.append(co)
values = values.append(so)
values = values.append(no)
values = values.append(pm)
values = values[['site','datetime','o3','co','so2','no2','pm10']]
values = values.rename(index=str, columns={"site":"place","datetime":"time"})
values['time'] =  pd.to_datetime(values['time'], format='%Y%m%dT%H%M-%S%f')
values['place'].replace('END OF FILE','',inplace=True)
values['time'].replace('', np.nan, inplace=True)
values.dropna(subset=['time'], inplace=True)
values['pm10'] = (values['pm10'])*(.01)
values['no2'] = (values['no2'])*(.1)
values['so2'] = (values['so2'])*(.1)
values['co'] = (values['co'])*(1)
values = values.fillna('')
```

```python
values = values.groupby(['time','place']).agg({'o3': 'first', 'co':'first' , 'so2':'first','no2':'first','pm10':'first'}).reset_index()
values = pd.DataFrame(data=values)

values['time'] = values['time'].dt.strftime('%m/%d/%y %H:%M')
```

```python
values.to_csv('data/values.csv')
```
